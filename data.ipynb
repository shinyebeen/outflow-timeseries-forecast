```python
import pandas as pd

df1 = pd.read_csv('data/마곡_목련.csv')
df2 = pd.read_csv('data/마곡_삼화그린.csv')
df3 = pd.read_csv('data/마곡_수성주공3.csv')
df4 = pd.read_csv('data/마곡_제일.csv')
df5 = pd.read_csv('data/마곡_한국원자력.csv')

name = ['목련', '삼화그린', '수성주공3', '제일', '한국원자력']
```

```python
df = pd.read_csv('C:/outflow/data/정읍데이터셋_소블록.csv')
df.head()
```

```python
df1.columns
```

```python
for name, data in zip(name, [df1, df2, df3, df4, df5]):
    data.drop(['상태', '최종지침<br>(㎥)', '단위사용량<br>(㎥/H)', 'RSSI<br>(dBm)', 'SNR<br>(db)', 'RSRQ', 'RSRP'], axis=1, inplace=True)
    data.rename(columns={'검침일시':'logTime',
                         '구간사용량<br>(㎥)':f'{name}_구간사용량(㎥)'}, inplace=True)
    data['logTime'] = pd.to_datetime(data['logTime'])
```

### 데이터 전처리

```python
df_ma.isna().sum()
```

```python
df_ma.set_index('logTime', inplace=True)
df_jang.set_index('logTime', inplace=True)
```

```python
def fill_missing_values(df):
    for col in df.columns[-5:]:
        df[col] = df[col].interpolate(method='time').fillna('ffill').fillna('bfill')

fill_missing_values(df_ma)
fill_missing_values(df_jang)
```

```python
df_ma.reset_index(inplace=True)
df_jang.reset_index(inplace=True)
```
